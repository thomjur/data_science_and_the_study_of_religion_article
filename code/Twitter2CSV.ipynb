```python
import json
import pandas as pd
```

# Functions

```python
def getTextFromTweet(tweet_object):
  '''
  function returning full text from tweet object
  '''
  if "extended_tweet" in tweet_object.keys():
    return tweet_object["extended_tweet"]["full_text"]
  elif "text" in tweet_object:
    return tweet_object["text"]
```

```python
def createCSVFromTweetList(tweet_list, file_name="default.csv"):
  '''
  function to create & save a csv from a list with tweet objects
  '''
  text_list = list()

  for tweet_object in tweet_list:
    text_list.append(getTextFromTweet(tweet_object))

  df = pd.DataFrame(text_list, columns=["text"])
  df.to_csv(f"drive/MyDrive/Colab Notebooks/{file_name}")
```

# Loading Tweets

Loading Tweets from File with JSON Tweet Objects

```python
tweet_list = list()

with open("drive/MyDrive/Colab Notebooks/Data/Twitter/Religion/tweet_religion.txt", "r") as f:
  for line in f.readlines():
    tweet_list.append(json.loads(line))
```

# Creating & Saving CSV

```python
createCSVFromTweetList(tweet_list)
```
