# Functions

```python
def get_tweets(q,max_results=10):
  '''
  function to get tweets from TODAY !!!
  '''
  # get tweets
  tweets = client.search_recent_tweets(q, max_results=max_results, tweet_fields=['text','context_annotations', 'id', 'entities'], start_time=datetime.datetime(year=time.localtime()[0], month=time.localtime()[1], day=time.localtime()[2]))

  # initialize dict
  dict_ = dict()

  # loop over tweets
  for idx, twt in enumerate(tweets.data):
    # get hashtags; empty string if no hashtags in tweet
    hashtags = get_hashtags(twt)

    # save information to dict and prepare for annotation
    dict_[idx] = dict()
    dict_[idx]['id'] = twt.id
    dict_[idx]['text'] = twt.text
    dict_[idx]['hashtags'] = hashtags
    dict_[idx]['ron'] = ''
    dict_[idx]['religion'] = ''
    dict_[idx]['social_spheres'] = ''
    dict_[idx]['sentiment'] = ''

    # create df    
    df = pd.DataFrame.from_dict(dict_, orient='index')

    # save as csv
    filename = 'twitter_rel_'+str(time.localtime()[0])+'-'+str(time.localtime()[1])+'-'+str(time.localtime()[2])+'.csv'
    df.to_csv(filename)
```

```python
def get_hashtags(twt):
  '''
  get hashtags from Twitter v2 API - if no hashtags return empty string
  '''
  # init empty list
  hashtag_list = list()
  
  # loop over hashtags if in tweet
  if twt['entities'] is not None:
    for key, value in twt['entities'].items():
        if key == 'hashtags':
            for hashtag in value:
                hashtag_list.append(hashtag['tag'])
  
  # return list as str or empty str
  if hashtag_list:
    return ' '.join(hashtag_list)
  else:
    return ''
```
